```python
df = pd.read_excel("/content/Final_Cleaned_English_Hindi_Dataset.xlsx")
```

```python
print(df.head)
```

```python
sample_df = df.sample(200, random_state=42)

english_sentences = sample_df["Source"].tolist()
reference_hindi = sample_df["Translation"].tolist()
```

```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from huggingface_hub import login

try:
    login()
except Exception as e:
    print(f"Could not log in to Hugging Face Hub: {e}")

model_name = "Helsinki-NLP/opus-mt-en-hi"

# Load tokenizer and model directly
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = AutoModelForSeq2SeqLM.from_pretrained(model_name)

# Create a translation function
def translate_text(text):
    inputs = tokenizer(text, return_tensors="pt", padding=True, truncation=True)
    outputs = model.generate(**inputs)
    translated_text = tokenizer.decode(outputs[0], skip_special_tokens=True)
    return translated_text

print(f"Model '{model_name}' loaded successfully and translation function created.")
```

```python
generated_hindi = []

for sentence in english_sentences:
    translated = translate_text(sentence)
    generated_hindi.append(translated)

print(generated_hindi[:5])
```

```python
from sacrebleu import corpus_bleu, corpus_chrf, corpus_ter

bleu = corpus_bleu(generated_hindi, [reference_hindi])
chrf = corpus_chrf(generated_hindi, [reference_hindi])
ter = corpus_ter(generated_hindi, [reference_hindi])

print("BLEU:", bleu.score)
print("CHRF:", chrf.score)
print("TER:", ter.score)
```
